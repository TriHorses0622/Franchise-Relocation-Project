```python
import pandas as pd
```

```python
census_data=pd.read_csv('Resources/census_data.csv')
venues_data=pd.read_csv('Resources/msv.csv')
david_work=pd.read_csv('Output/David_work.csv')
```

```python
census_data=census_data.drop(columns=['Unnamed: 0'])
census_data=census_data.rename(columns={'respop72018':'POPULATION'})
census_data.head()
```

```python
venues_data['CITY']=venues_data['CITY'].str.lower()
venues_data['CITY']=venues_data['CITY'].str.replace(' ','')
venues_data['CITY']=venues_data['CITY'].str.replace('saint','st.')
venues_data.head()
```

```python
merge_table = pd.merge(census_data, venues_data, on=['CITY','STATE'],how='right')
merge_table = merge_table.dropna()
merge_table.head()
```

```python
merge_grouped=merge_table.groupby(['CITY','STATE'])

population=merge_grouped['POPULATION'].mean()
venuesCount=merge_grouped['VENUEID'].count()

shrinkedDF=pd.DataFrame({
    'Population':population,
    'Venues Count':venuesCount
})

shrinkedDF=shrinkedDF.sort_values(['Population'],ascending=False)
shrinkedDF=shrinkedDF.reset_index()
shrinkedDF.head()
```

```python
shrinkedDF=shrinkedDF.rename(columns={'CITY':'City'})
shrinkedDF
shrinkedDF.to_csv('Output/newshrink.csv')
```

```python
david_work['City']=david_work['City'].str.lower()
david_work['City']=david_work['City'].str.replace(' ','')
david_work.head()
```

```python
new_merge_table=pd.merge(shrinkedDF,david_work,on=['City'],how='right')
new_merge_table.head()
```

```python
new_merge_table.to_csv('output/new_merge_table.csv')
```
